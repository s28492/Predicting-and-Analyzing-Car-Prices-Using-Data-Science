```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('wyniki_tuningu.csv')
df['MSE'] = -df['mean_test_score']
df['RMSE'] = np.sqrt(df['MSE'])
df.info()
```

```python
df.sort_values(by=['RMSE'], ascending=True, inplace=True)
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

params_to_plot = ['n_estimators', 'max_depth', 'learning_rate',
                 'subsample', 'colsample_bytree', 'gamma',
                 'reg_alpha', 'reg_lambda']

plt.figure(figsize=(20, 15))
for i, param in enumerate(params_to_plot, 1):
    plt.subplot(3, 3, i)
    sns.scatterplot(x=param, y='RMSE', data=df, alpha=0.7, hue='gamma')
    plt.title(f'{param} vs RMSE')
    plt.yscale('log')
plt.tight_layout()
plt.show()
```

```python
from pandas.plotting import parallel_coordinates

top_20 = df.loc[:20]
plt.figure(figsize=(15, 10))
parallel_coordinates(top_20[['n_estimators', 'max_depth', 'learning_rate',
                           'subsample', 'RMSE']], 'RMSE', colormap='viridis')
plt.title('Równoległe Współrzędne dla Top Kombinacji')
plt.xticks(rotation=45)
plt.show()
top_20
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['RMSE'], kde=True, bins=20)
plt.title('Rozkład wartości RMSE')
plt.xlabel('RMSE')
plt.ylabel('Liczba prób')
plt.show()
```

```python
pivot_table = df.pivot_table(values='RMSE',
                            index='learning_rate',
                            columns='max_depth')

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='coolwarm', annot=True, fmt=".1f")
plt.title('Interakcja learning_rate i max_depth')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='n_estimators', y='RMSE', data=df)
plt.title('Stabilność wyników dla różnych liczb drzew')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='mean_fit_time', y='RMSE', data=df,
               hue='max_depth', size='n_estimators')
plt.title('Relacja czasu treningu do jakości modelu')
plt.xlabel('Średni czas treningu (s)')
plt.ylabel('RMSE')
plt.show()
```
